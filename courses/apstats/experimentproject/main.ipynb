```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# importing dataset
data = pd.read_csv('data/advMLdata.csv')
data.head()
```

```python
data.boxplot(by='Treatment', column='Confidence of Model in Predicting True Label',vert=0)
plt.title('Confidence in Predicting True Label by Model')
plt.suptitle('')
plt.xlabel('Confidence in Predicting True Label (Probability)')
plt.show()
```

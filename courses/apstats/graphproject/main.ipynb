```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# importing dataset
data = pd.read_csv('data/responses_cleaned.csv')
data.head()
```

```python
data.boxplot(by='Favorite Color', column='Minutes to Get Ready',vert=0)
plt.title('Student Time to Get Ready by Favorite Color')
plt.suptitle('')
plt.xlabel('Time to Get Ready (Min)')
plt.show()
```
